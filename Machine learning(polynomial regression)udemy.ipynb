```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#read csv file and importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
#training the linear regression model on the whole dataset.
from sklearn.linear_model import LinearRegression
lin_reg= LinearRegression()
lin_reg.fit(X,y)
```

```python
#training the polynomial model on the whole dataset.
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures (degree= 2)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
#Visualising the linear regression result
plt.scatter(X, y, color= 'red')
plt.plot (X, lin_reg.predict(X), color = 'blue')
plt.title('Truth vs Bluff (Linear regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising the polynomial regression result
plt.scatter(X, y, color= 'red')
plt.plot (X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth vs Bluff (Polynomial regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#if we increase the degree the result will be more precise
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures (degree= 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
#Visualising the polynomial regression result(for higher resolution and smoother curve)
X_grid = np.arange(min(X), max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X, y, color= 'red')
plt.plot (X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth vs Bluff (Polynomial regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#predicting a new result with linear regression
lin_reg.predict([[6.5]])
```

```python
#predicting a new result with polynomial regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
